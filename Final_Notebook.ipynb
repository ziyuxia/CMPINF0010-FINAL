Introduction:

Our project for the best neighborhood in Pittsburgh focused on what makes a neighborhood most relaxing. For our metrics, we decided there were three factors on what makes a neighborhood relaxing: the number of dogs (determined by the dog licenses registered to that neighborhood), the number of parks, and the number of playgrounds. We decided on these metrics because they are not determined by momentary or social constraints; anyone can pet a dog, go for a walk, and just use a playground. We previously considered other metrics like number of restaurants in a neighborhood, but decided against it in favor of other more readily available amenities.  



Metric:

For our metrics, we ranked the neighborhoods in each metric (dogs, parks, and playgrounds) relatively to each other. In each metric, we just simply used a bar chart to show how the neighborhoods ranked with each other; the charts were truncated to 15 or so neighborhoods for better data visualization and because the number of dogs, parks, or playgrounds dropped off at that point. We used the following data sets from the  Western Pennsylvania Regional Data Center(WPRDC):



    1) "City of Pittsburgh Playgrounds", https://data.wprdc.org/dataset/playgrounds, The number of playgrounds in a given neighborhood
    Note: The "Allegheny County Zip Code Boundaries", https://data.wprdc.org/dataset/allegheny-county-zip-code-boundaries2, data set was used with the Allleghent County Dog Licenses data set to associate each zip code to a certain neighborhood. 
    2) "Allegheny County Dog Licenses", https://data.wprdc.org/dataset/allegheny-county-dog-licenses, the dog licenses data set per zip code
    3) "City of Pittsburgh Parks", https://data.wprdc.org/dataset/parks, the number of parks per neighborhood dataset 
    

Best Neighborhood:

Our criteria for the best neighborhood in Pittsburgh required the neighborhood to maximize all three metrics. In our search, we saw that many neighborhoods lined with the extremes; the neighborhoods with the most dogs did not have the most parks or playgrounds, the neighborhoods with the most parks did not have the most playgrounds or dogs, and vice versa. The only exception was Squirrel Hill South; the neighborhood that consistently had a high amount of parks, playgrounds, and dogs.
